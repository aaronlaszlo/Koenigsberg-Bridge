```python
print("Now we check if there is a eulerian path (does not have to return)")
print("+---------------CASE 1a:--------------+")
print("Euler Path?: ", has_eulerian_path(A1))
print("+---------------CASE 1b:--------------+")
print("Euler Path?: ", has_eulerian_path(A2))
print("+---------------CASE 2a:--------------+")
print("Euler Path?: ", has_eulerian_path(B1))
print("+---------------CASE 2b:--------------+")
print("Euler Path?: ", has_eulerian_path(B2))
print("+---------------CASE 3a:--------------+")
print("Euler Path?: ", has_eulerian_path(C1))
print("+---------------CASE 3b:--------------+")
print("Euler Path?: ", has_eulerian_path(C2))


print("Now we check if journey is a eulerian (does have to return)")
print("+---------------CASE 1a:--------------+")
print("Euler Path?: ", is_eulerian(A1))
print("+---------------CASE 1b:--------------+")
print("Euler Path?: ", is_eulerian(A2))
print("+---------------CASE 2a:--------------+")
print("Euler Path?: ", is_eulerian(B1))
print("+---------------CASE 2b:--------------+")
print("Euler Path?: ", is_eulerian(B2))
print("+---------------CASE 3a:--------------+")
print("Euler Path?: ", is_eulerian(C1))
print("+---------------CASE 3b:--------------+")
print("Euler Path?: ", is_eulerian(C2))
```

```python
graphs = [A1, A2, B1, B2, C1, C2]
names = ["A1", "A2", "B1", "B2", "C1", "C2"]

print("Paths without returning:")
for idx, path in enumerate(graphs):
    
    if has_eulerian_path(path):
        
            print("+---------Path for ", names[idx], "---------+")
            print([u for u, v in nx.eulerian_path(path)])

print("\n", "Paths with returning:")

for idx, path in enumerate(graphs):
    
    if is_eulerian(path):
        
            print("+---------Path for ", names[idx], "---------+")
            print([u for u, v in nx.eulerian_circuit(path)])
```
